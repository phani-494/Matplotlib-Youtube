## Stacked bar chart with example in matplotlib| horizontal and vertical stacked bar chart

```python
# here we will consider the placements of the students for four years.
years=["2017","2018","2019","2020"] # we should give here al string values .
placements=[256,134,502,446] # placements of entire college
plt.bar(years,placements,width=0.4)
plt.title("Placement in college")
plt.show()
```

```python
# we need to divide these placements in branch wise also. 
# here if we observe clearly, we can see that all are overlapping each other.
# we can also give limits o the x-axis and y-axis as plt.ylim(0,100)
years=["2017","2018","2019","2020"] 
placements=[256,134,502,446] 
cse=[75,80,100,223]
it=[46,30,200,100]
ece=[30,20,72,62]
plt.bar(years,ece,width=0.4,color="red")
plt.bar(years,it,width=0.4,color="green")
plt.bar(years,cse,width=0.4)
plt.title("Placement in college")
plt.ylim(0,550)
plt.show()
```

```python
years=["2017","2018","2019","2020"] 
placements=[256,134,302,446] 
cse=[75,80,50,23]
it=[46,30,19,10]
ece=[30,20,72,62]
cse_start_value=[it[i]+ ece[i] for i in range(len(ece)) ]
print(cse_start_value)
plt.bar(years,ece,width=0.4,color="red")
plt.bar(years,it,bottom=ece,width=0.4,color="green")
plt.bar(years,cse,bottom=cse_start_value,width=0.4,color="k") # after the ece and it, we have to start the cse values here
plt.title("Placement in college")
plt.ylim(0,160)
plt.show()
```

```python
# similarly ,if we use barh, we need to give the paramter as left and 
years=["2017","2018","2019","2020"] 
placements=[256,134,302,446] 
cse=[75,80,50,23]
it=[46,30,19,10]
ece=[30,20,72,62]
cse_start_value=[it[i]+ ece[i] for i in range(len(ece)) ]
print(cse_start_value)
plt.barh(years,ece,height=0.4,color="red")
plt.barh(years,it,left=ece,height=0.4,color="green")
plt.barh(years,cse,left=cse_start_value,height=0.4,color="k") # after the ece and it, we have to start the cse values here
plt.title("Placement in college")
plt.xlabel("placements")
plt.show()
```

```python
# suppose if you need to add one more branch eee also for this.
years=["2017","2018","2019","2020"] 
placements=[256,134,302,446] 
cse=[75,80,50,23]
it=[46,30,19,10]
ece=[30,20,72,62]
cse_start_value=[it[i]+ ece[i] for i in range(len(ece)) ]
print(cse_start_value)
eee=[100,40,50,40]
eee_start_value=[cse[i]+ cse_start_value[i] for i in range(len(ece)) ]
plt.barh(years,ece,height=0.4,color="red")
plt.barh(years,it,left=ece,height=0.4,color="green")
plt.barh(years,cse,left=cse_start_value,height=0.4,color="k") # after the ece and it, we have to start the cse values here
plt.barh(years,eee,left=eee_start_value,height=0.4,color="y") # after the ece and it, we have to start the cse values here

plt.title("Placement in college")
plt.xlabel("placements")
plt.show()
```

```python
years=["2017","2018","2019","2020"] 
placements=[256,134,302,446] 
cse=[75,80,50,23]
it=[46,30,19,10]
ece=[30,20,72,62]
cse_start_value=[it[i]+ ece[i] for i in range(len(ece)) ]
print(cse_start_value)
eee=[100,40,50,40]
eee_start_value=[cse[i]+ cse_start_value[i] for i in range(len(ece)) ]
plt.barh(years,ece,height=0.4,color="red",label="ece")
plt.barh(years,it,left=ece,height=0.4,color="green",label="it")
plt.barh(years,cse,left=cse_start_value,height=0.4,color="k",label="cse") 
plt.barh(years,eee,left=eee_start_value,height=0.4,color="y",label="eee") 
plt.legend()
plt.title("Placement in college")
plt.xlabel("placements")
plt.show()
```

# lec 14
## Grouped bar chart || Multiple bar chart with example

```python
# we can also display the above stacked representation individually without any overlapping or bottom.
import numpy as np
years=["2017","2018","2019","2020"] 
placements=[256,134,302,446] 
cse=[75,80,50,23]
it=[46,30,19,10]
ece=[30,20,72,62]
eee=[100,40,50,40]
w=0.2 # width of each bar is 0.2
# here we will plot each department individually  instead of the stacked way.
# on the x-axis instead of the years,we need to give the number values (total how many years we are representing)
cse_bar=np.arange(len(years)) # total 4 bars
plt.bar(cse_bar,cse,width=w,label="cse")
plt.legend()


plt.title("Placement")
plt.show()
```

```python
# now we have to represent the it placement just after the cse.
# since the first bar is starting at 0, the next bar should be started after the 0.2 

it_bar=[i+w for i in cse_bar]
plt.bar(cse_bar,cse,width=w,label="cse")
plt.bar(it_bar,it,width=w,label="it")
plt.legend()


plt.title("Placement")
plt.show()
```

```python
ece_bar=[i+w for i in it_bar]
plt.bar(cse_bar,cse,width=w,label="cse")
plt.bar(it_bar,it,width=w,label="it")
plt.bar(ece_bar,ece,width=w,label="ece")
plt.legend()


plt.title("Placement")
plt.show()

# in the similar way ,we can plot the eee also.

```

```python
# once u done this, now we can change the x-values to  whatever the values we want by using the x-ticks.
ece_bar=[i+w for i in it_bar]
plt.bar(cse_bar,cse,width=w,label="cse")
plt.bar(it_bar,it,width=w,label="it")
plt.bar(ece_bar,ece,width=w,label="ece")
plt.legend()
plt.xticks(cse_bar,years) # instead of the cse bar just place the years values.

plt.title("Placement")
plt.show()
```

```python
# if u want to move the tick value then just add it.
plt.bar(cse_bar,cse,width=w,label="cse")
plt.bar(it_bar,it,width=w,label="it")
plt.bar(ece_bar,ece,width=w,label="ece")
plt.legend()
plt.xticks(cse_bar+w,years) # instead of the cse bar just place the years values.
plt.figure(figsize=(100,100))
plt.title("Placement")
plt.show()
```

## Pie chart in matplotlib | Pie Graph

```python
# let us have the grades and the no of students who got the grades in pie chart.
import matplotlib.pyplot as plt
grades=["Excellent","good","avg","poor"]
students=[50,60,40,70]
plt.pie(students)
plt.show()
```

```python
import matplotlib.pyplot as plt
grades=["Excellent","good","avg","poor"]
students=[50,60,40,70]
plt.pie(students,labels=grades,startangle=90) # start at y-axis [90 degrees] 
plt.show()
```

```python
# if u want to explode some portion out from the pie chart,then we need to give explode parameter
# just we want excellent outside from the cirlce, we can give some value for it.
# by default explode will have the default values as zero.
plt.pie(students,labels=grades,explode=[0.2,0,0,0.1])  # some points distance 
plt.show()
```

```python
# shadow is present in pie charts also.
plt.pie(students,labels=grades,startangle=90,shadow=True,explode=[0.2,0,0,0.1]) # start at y-axis [90 degrees] 
plt.show()
```

```python
plt.figure(figsize=(8,10)) # we can give the size of the figure we want here.
plt.pie(students,labels=grades,colors=["r","y","k","g"]) 
plt.legend(loc="best",title="performances")
plt.show()
```

```python
# we can represent the percentage of the each portion in the pie chart.
# autopct is the parameter used here and %2.2f%% means 2 digits after decimal and 2 digits after decimal point.
plt.pie(students,labels=grades,colors=["r","y","k","g"],autopct="%2.2f%%") 
#plt.legend(loc="best",title="performances")
plt.show()
```

## Histogram chart in matplotlib | Histogram graph | histogram plot

```python
marks=[10,20,30,34,23,67,89,23,67,55]
grades_intervals=[0,35,60,100] # here the intervals are 0-35, 35-60,60-100
plt.hist(marks,grades_intervals)
plt.title("Histogram ")
plt.show()
```

```python
# the above graph shows that from 0-35 we have 6 students,
# from 35-60 one student and from 60-100 three are present.
```

```python
# in order to not get confused, we can use xticks() and give the range values.
marks=[10,20,30,34,23,67,89,23,67,55]
grades_intervals=[0,35,60,100] # here the intervals are 0-35, 35-60,60-100
plt.hist(marks,grades_intervals)
plt.title("Histogram ")
plt.xticks([0,35,60,100])
plt.show()
```

```python
# there are different types of histogram.
# we can represent them as histtype="bar"  or "step" or "stepfilled" 
#rwidth also present for histogram

marks=[10,20,30,34,23,67,89,23,67,55]
grades_intervals=[0,35,60,100] # here the intervals are 0-35, 35-60,60-100
plt.hist(marks,grades_intervals,rwidth=0.5)
plt.title("Histogram ")
plt.xticks([0,35,60,100])
plt.show()
```

```python
marks=[10,20,30,34,23,67,89,23,67,55]
grades_intervals=[0,35,60,100] # here the intervals are 0-35, 35-60,60-100
plt.hist(marks,grades_intervals,rwidth=0.5,histtype="step")
plt.title("Histogram ")
plt.xticks([0,35,60,100])
plt.show()
```

```python
marks=[10,20,30,34,23,67,89,23,67,55]
grades_intervals=[0,35,60,100] # here the intervals are 0-35, 35-60,60-100
plt.hist(marks,grades_intervals,rwidth=0.5,histtype="stepfilled",facecolor="green")
plt.title("Histogram ")
plt.xticks([0,35,60,100])
plt.show()
```

## xticks(),yticks(),xlim(),ylim() in matplotlib

```python
# let us draw a simple line plot here
x=[1,2,3,4,5]
y=[3,5,6,10,8]
plt.plot(x,y)
plt.show()
```

```python
# suppose for the x-axis,instead of the default values we want to have some other names.
# we need to use xticks() here. if we observe x-axis,a small vertical line is present. it is called as tick.
x=[1,2,4,3,5]
y=[3,5,6,10,8]
plt.plot(x,y)
plt.xticks(x,["a","b","c","d","e"]) # based on the x-values only, these are represented.
plt.yticks(y,["a","b","c","d","e"])
plt.show()
```

```python
x=[1,2,4,3,5]
y=[3,5,6,10,8]
plt.plot(x,y)
plt.xticks(x,["a","b","c","d","e"]) # based on the x-values only, these are represented.
plt.yticks(y,["0-100","100-200","200-300","300-400","400-500"])
plt.show()
```

```python
# to change the limit of the x-axis and y-axis, we can use the xlim and ylim.
x=[1,30,55,46,59]
y=[103,206,209,108,138]
plt.plot(x,y)
plt.xlim(0,100) # ([10,100])
plt.ylim(100,300)
plt.show()
```
