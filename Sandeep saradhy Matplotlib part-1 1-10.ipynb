## Simple plot in MATPLOTLIB | Line plot in Matplolib

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
# suppose in the plot if we give only one list value,by default it will consider as y-axis values here.
# and for x-axis,it will defaultly consider as starting from zero.
plt.plot([10,20,30]) # see here the x-axis is treated as [0,1,2]
plt.show()
```

```python
# plt.xlabel(name) gives a label for x-axis.
# plt.ylabel(name) gives a label for y-axis.
# plt.title(name) displays the title for the chart.
plt.plot([10,20,30])
plt.xlabel("Numbers")
plt.ylabel("Marks")
plt.title("My chart 1")
plt.show()
```

# lec 4
## Lines(style, width,color) in matplotlib| Line properties in matplotlib

```python
The linestyle can be a solid line(default) as ls="solid" 
ls="dotted" # gives dotted line
"dashed" line,"dashdot" line as -.-.-.
"None" doesn't appear.

linewidth is the width of the line. lw=10
linecolor is given as c="r".
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=x**2
plt.plot(x,y,ls="dotted",lw=3,c="k")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=x**2
plt.plot(x,y,ls="dashdot",lw=2,c="b")
plt.show()
```

```python
# if we give None then no line will be displayed.
# In such cases it is better to use the marker there.
x=np.arange(0,10,2)
y=x**2
plt.plot(x,y,ls="None",lw=2,c="b")
plt.show()
```

```python
x=np.arange(0,10,2)
y=x**2
plt.plot(x,y,ls="None",lw=2,c="b",marker="d",mec="r")
plt.show()
```

```python
# we can also represent multiple lines in a single graph .
x=np.arange(0,10,2)
y=x**2
y1=x**3
plt.plot(x,y,lw=2,c="b",marker="d",mec="r")
plt.plot(x,y1,lw=2,c="g",marker="o",mec="r")
plt.show()
```

```python
plt.plot(x,y,x,y1) # 
we can also give it in a single line here
plt.show()
```

# lec 5
## Font properties on title ,labels in matplotlib

```python
x=np.arange(0,10,2)
y=x**2
y1=x**3
plt.plot(x,y,lw=2,c="b",marker="d",mec="r")
plt.plot(x,y1,lw=2,c="g",marker="o",mec="r")
f1={"family":"italic","size":20,"color":"r"}
f2={"family":"arial","size":20,"color":"b"}
f3={"family":"calibri"}
plt.xlabel("X values are here",fontdict=f1)
plt.ylabel("X square values are here",fontdict=f2)
plt.title("Font styles graph",fontdict=f3)
plt.show()
```

## Grid function with the line properties in matplotlib | Grid lines

```python
x=np.arange(0,10,2)
y=x**2
y1=x**3
plt.plot(x,y)
plt.xlabel("X values are here")
plt.ylabel("X square values are here")
plt.title("Font styles graph")
plt.grid()
plt.show()
```

```python
x=np.arange(0,10,2)
y=x**2
y1=x**3
plt.plot(x,y)
plt.xlabel("X values are here")
plt.ylabel("X square values are here")
plt.title("Font styles graph")
plt.grid(axis="x")
plt.show()
```

# lec 7
## Subplots in Matplotlib

```python
x1=[1,2,3,4,5]
y1=[2,4,4,5,6]
x2=[2,5,4,5,6]
y2=[2,3,6,5,6]
x3=[2,3,4,7,6]
y3=[2,3,9,5,6]
x4=[2,1,4,5,6]
y4=[2,3,4,8,6]
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("index (2,2,1)")
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("index (2,2,2)")
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("index (2,2,3)")
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("index (2,2,4)")
plt.suptitle("All 4 sub plots")
plt.show()
```

# lec 8
## Legend function in matplotlib

```python
x=np.arange(0,13,2)
y=x**2
y1=x**3
plt.plot(x,y,x,y1)
plt.legend(["sqaures","cubes"],loc=2,framealpha=0.5,facecolor="red",edgecolor="k",fancybox=False)
plt.show()
```

```python
x=np.arange(0,13,2)
y=x**2
y1=x**3
plt.plot(x,y,label="sqaures")
plt.plot(x,y1,label="cubes")
plt.legend() # here the labels are automatically detected when there are no parameters in legend.
plt.show()
```

```python
x=np.arange(0,13,2)
y=x**2
y1=x**3
plt.plot(x,y,x,y1)
plt.legend(["sqaures","cubes"],loc="center",framealpha=0.5,facecolor="red",edgecolor="k",fancybox=False)
plt.show()
```

```python
x=np.arange(0,13,2)
y=x**2
y1=x**3
plt.plot(x,y,x,y1)
plt.legend(["sqaures","cubes"],fontsize=15,loc="center",framealpha=0,facecolor="red",edgecolor="k")
plt.show()
```

# lec 9
## Creating scatter plot in matplotlib

```python
import random
x=np.arange(0,13,2)
y=x**2
print(y)
random.shuffle(y)
print(x,y)
plt.scatter(x,y)
plt.show()
```

```python
import random
x=np.arange(0,13,2)
y=x**2
print(y)
random.shuffle(y) # doing shuffle so that we can understand the shuffle very neatly.
print(x,y)
plt.scatter(x,y,color="r",marker="^") # gives the triangle shape here.
plt.show()
```

```python
x=np.arange(1,13,2)
y=x**2
y1=x**3
print(y)
random.shuffle(y)
random.shuffle(y1)
print(x,y)
plt.scatter(x,y,marker="*",s=100)

plt.scatter(x,y1,color="red")
plt.show()
```

```python
x=np.arange(1,13,2)
y=x**2
y1=x**3
print(y)
random.shuffle(y)
random.shuffle(y1)
print(x,y)
plt.scatter(x,y,marker="*",s=100)

plt.scatter(x,y1,color="red",linewidth=5)
plt.show()
```

## CMAP and  COLORBAR (scatter plot)

#### representing the individual points with different color and different size

```python
x=np.arange(1,13,2)
y=x**2
random.shuffle(y)
random.shuffle(y1)
colors=["r","g","b","k","y","r"]
print(x,y)
plt.scatter(x,y,marker="*",s=100,color=colors) # we can also give it as c=colors
```

```python
x=np.arange(1,13,2)
y=x**2
random.shuffle(y)
random.shuffle(y1)
sizes=[10,120,140,160,180,200]
colors=["r","g","b","k","y","r"]
print(x,y)
plt.scatter(x,y,marker="*",color=colors,s=sizes)
```

```python
# colormaps: representing the colors in terms of integers from 0 to 90 having different colors.
# we have lot of different cmaps here.
x=np.arange(1,13,2)
y=x**2
random.shuffle(y)
random.shuffle(y1)
sizes=[10,120,140,160,180,200]
colors=[20,25,67,80,30,90] # instead of names we will give integers here  simply.
plt.scatter(x,y,marker="*",c=colors,cmap="viridis",s=sizes)
plt.colorbar() # here we can see the color bar range # starts with minimum and ends with maximum in bar.
```
